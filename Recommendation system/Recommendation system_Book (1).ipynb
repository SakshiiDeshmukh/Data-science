### Recommendation System
A recommendation system is a type of information filtering system that suggests items to users based on their preferences, past behavior, or similarity to other users. The goal of a recommendation system is to provide personalized and relevant recommendations to users, helping them discover new items or improve their overall experience.

Recommendation Techniques:

* Collaborative Filtering: This technique recommends items based on the similarity between users. It finds users who have similar preferences and recommends items that those similar users have liked or preferred. The idea is that if users have similar tastes, they are likely to have similar preferences for items.
* Content-Based Filtering: This technique recommends items based on the similarity between items themselves. It analyzes the characteristics or attributes of items and recommends similar items to what the user has liked or interacted with in the past. It focuses on the content or features of items, such as genre, director, or keywords.
* Hybrid Approaches: These approaches combine collaborative filtering and content-based filtering to provide more accurate recommendations. They leverage both user preferences and item features to generate personalized recommendations. By combining the strengths of both techniques, hybrid approaches aim to improve the quality and relevance of recommendations.

### Using cosine function to find the more similar books based on the lower cosine angle between the vectors

The function "pairwise_distances" calculates the distances or similarities between pairs of vectors. When you subtract the result from 1, you are converting the similarity values to dissimilarity values.


By subtracting the pairwise distances from 1, you are essentially transforming the values to represent dissimilarity. This is a common approach, where 1 indicates perfect similarity and O represents perfect dissimilarity.

```python
c_sim = 1 - pairwise_distances( book_df.values,metric='cosine')
c_sim.shape
```

### Creating a user defined function, 

That will get the book title and the cosine similarity as input and return a list of books thats are similar to the book  that was given as input argument to the function

```python
def get_recommendation(book_title,cosine_sim=c_sim):
    
    idx= indices[book_title]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    sim_scores = sorted(sim_scores,key=lambda x:x[1],reverse =True)
    
    sim_scores  = sim_scores[0:11]
    
    book_id    = [i[0] for i in sim_scores]
    
    return books['book_title'].iloc[book_id]
```

1. enumerate () allows you to iterate over a sequence (such as a list, tuple, or string) while keeping track of the index of each element.


2. The lambda function is defined as lambda x: x[1]. It takes a single argument x and returns x[1], which is the second element of the tuple x.


3. i[0] for i in sim_score It creates a new list called books_indices by extracting the first element of each tuple in the sim_scores list.
